## Once segmented an image, we want to identify objects.Takes in a binary image,return a close fitting boundary(contour) around evry object. It is called an iterative energy reducton algo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img=cv2.imread('obj.png',1)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
thresh=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
```

```python
_,contours,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
img2=img.copy()
```

```python
index=-1  # draw all the contours as -1
```

```python
thickness=4
```

```python
color=(255,255,0)
```

```python
cv2.drawContours(img2,contours,index,color,thickness)
```

```python
cv2.imshow("Contours",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("ContouredObj.jpg",img2)  # see u could change the extension so easily
```

## Area , Perimeter, Center of CUrvature

```python
objects=np.zeros([img.shape[0],img.shape[1],3],'uint8')
count=0
```

```python
for c in contours:
    cv2.drawContours(objects,[c],-1,color,-1) # last -1 means fill the complete
    area=cv2.contourArea(c)
    perimeter=cv2.arcLength(c,True)
    M=cv2.moments(c)
    
    if((M['m00'])!=0):
        cx=int(M['m10']/M['m00'])
        cy=int(M['m01']/M['m00'])
        cv2.circle(objects,(cx,cy),4,(0,0,255),-1)
        
    else:
        print("division by zero")
    
    print("Area : {} , perimeter: {}".format(area,perimeter))
    
    
cv2.imshow("Contours",objects)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("Objects.jpg",objects)
```

### Multiple centroids detected because of the complexity of objects

```python
filtered=[]
for c in contours:
    if cv2.contourArea(c)<1000:
        continue
    filtered.append(c)
```

```python
import random
for c in filtered:
    col=(random.randint(0,255),random.randint(0,255),random.randint(0,255))
    cv2.drawContours(objects,[c],-1,col,-1)
    area=cv2.contourArea(c)
    perimeter=cv2.arcLength(c,True)
    print(area,perimeter)
cv2.imshow("contours",objects)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
