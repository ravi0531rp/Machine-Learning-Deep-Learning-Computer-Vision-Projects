```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
bfly=cv2.imread("bfly.jpg",1)
```

```python
gray=cv2.cvtColor(bfly,cv2.COLOR_RGB2GRAY)
```

```python
cv2.imwrite("gray.jpg",gray)
```

```python
plt.imshow(gray)
plt.show()
```

## Add an additional channel. Say, the transparency or alpha channel

```python
b=bfly[:,:,0]
```

```python
g=bfly[:,:,1]
```

```python
r=bfly[:,:,2]
```

### If u want to make the non green parts of the image transparent,we could pass in the g channel as the additional one, where a high value will show highly opaque and reverse.

```python
rgba=cv2.merge((b,g,r,g))
```

```python
cv2.imwrite("rgba.png",rgba)  # jpg doesn't support imng transparency
```

```python
plt.imshow(rgba)
plt.show()
```

## Gaussian Blur, dilation and erosion

### Gaussian Blur

```python
noisy=cv2.imread('noisy.jpg')
```

```python
blur=cv2.GaussianBlur(noisy,(5,55),0)  # 5,55 = x and y bluuring axis
```

```python
cv2.imshow("Blurred",blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(noisy)
plt.show()
```

```python
plt.imshow(blur)
plt.show()
```

### Dilation and erosion filter
They work by sliding a small square kernel across the image

Dilation turns black to white and erosion the reverse

```python
kernel=np.ones((5,5),'uint8')
```

```python
dilate=cv2.dilate(noisy,kernel,iterations=1)
erode=cv2.erode(noisy,kernel,iterations=1)
```

```python
cv2.imshow("dilate",dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(dilate)
plt.show()
```

```python
cv2.imshow("erode",erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(erode)
plt.show()
```

## Scale and Rotate images

```python
img=cv2.imread('Road621.jpg',1)
```

```python
plt.imshow(img)
plt.show()
```

```python
imgHalf=cv2.resize(img,(0,0),fx=0.5,fy=0.5) # resize to 1/2 on both axis. (0,0) means that we won't give the number of pixels by hand
```

```python
imgStretch=cv2.resize(img,(600,600))
```

```python
imgStretchNear=cv2.resize(img,(600,600),interpolation=cv2.INTER_NEAREST)
```

```python
cv2.imshow("Half",imgHalf)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(imgHalf)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,3,1)
ax1.imshow(imgHalf)
ax2 = fig.add_subplot(3,3,2)
ax2.imshow(imgStretch)
ax3 = fig.add_subplot(3,3,3)
ax3.imshow(imgStretchNear)
ax4 = fig.add_subplot(3,3,4)
ax4.imshow(img)
```

### Rotate img via matrix tranform

```python
M=cv2.getRotationMatrix2D((0,0),-30,1)
```

```python
rotated=cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))
```

```python
cv2.imshow("ro",rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(rotated)
plt.show()
```
